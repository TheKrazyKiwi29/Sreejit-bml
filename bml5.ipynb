import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier
```

load the dataset

```python
from sklearn.datasets import load_diabetes
data = load_diabetes()

df = pd.DataFrame(data.data, columns=data.feature_names)
df['Outcome'] = (data.target > data.target.mean()).astype(int)  # Creating a binary target variable
```

explore the dataset

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df.shape
```

data normalization
min -max

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
df_normalized = pd.DataFrame(mm.fit_transform(df), columns=df.columns)
df_normalized
```

z-score standardisation

```python
mm = MinMaxScaler()
df_standardized = pd.DataFrame(mm.fit_transform(df), columns=df.columns)
df_standardized
```

splitting the dataset

```python
x = df_normalized.drop(columns=['Outcome'])  # Use df_normalized instead of df
y = df[['Outcome']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("Training size:", x_train.shape)
print("Testing size:", x_test.shape)
```

logistic regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
print("Test Score:", model.score(x_test, y_test))
print("Train Score:", model.score(x_train, y_train))
```

naive bayes (gaussian)

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
print("Test Score:", model.score(x_test, y_test))
print("Train Score:", model.score(x_train, y_train))
```

multinomial

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train, y_train)
print("Test Score:", model.score(x_test, y_test))
print("Train Score:", model.score(x_train, y_train))
```

bernouli

```python
from sklearn.naive_bayes import BernoulliNB
model = BernoulliNB()
model.fit(x_train, y_train)
print("Test Score:", model.score(x_test, y_test))
print("Train Score:", model.score(x_train, y_train))
```

classification report

```python
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```

accuracy score

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)
```

knn(manual)

```python
from collections import Counter

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

def knn_predict(x_train, y_train, X_test, k):
    predictions = []
    for test_sample in X_test.values:
        distances = np.sqrt(np.sum((x_train.values - test_sample)**2, axis=1))
        k_neighbors = np.argsort(distances)[:k]
        k_labels = [y_train.iloc[i, 0] for i in k_neighbors]
        most_common = Counter(k_labels).most_common(1)[0][0]
        predictions.append(most_common)
    return np.array(predictions)

k = 20
y_pred_manual = knn_predict(x_train, y_train, x_test, k)
accuracy_manual = np.mean(y_pred_manual == y_test['Outcome'].values)
print(f"Manual KNN Accuracy with k={k}: {accuracy_manual:.2%}")
```
